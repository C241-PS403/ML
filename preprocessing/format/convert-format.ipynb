```python
import os
from PIL import Image
import shutil
```

```python
def convert_and_save_as_jpeg(image_path, output_path):
    try:
        with Image.open(image_path) as img:
            base_name = os.path.basename(image_path)
            name, ext = os.path.splitext(base_name)
            jpeg_image_path = os.path.join(output_path, f"{name}.jpg")
            
            rgb_image = img.convert('RGB')
            rgb_image.save(jpeg_image_path, format='JPEG')
            print(f"Converted {image_path} to {jpeg_image_path}")
    except Exception as e:
        print(f"Error converting {image_path}: {e}")
```

```python
def process_directory(input_dir, output_dir):
    for root, dirs, files in os.walk(input_dir):
        for file in files:
            input_file_path = os.path.join(root, file)
            relative_path = os.path.relpath(input_file_path, input_dir)
            output_file_path = os.path.join(output_dir, relative_path)
            output_folder_path = os.path.dirname(output_file_path)
            
            if not os.path.exists(output_folder_path):
                os.makedirs(output_folder_path)
            
            if file.lower().endswith(('.png', '.bmp')):
                convert_and_save_as_jpeg(input_file_path, output_folder_path)
            elif file.lower().endswith('.jpg'):
                shutil.copy2(input_file_path, output_folder_path)
                print(f"Copied {input_file_path} to {output_folder_path}")
```

```python
input_dir = 'C:\\Bangkit\\ML\\code\\preprocessing\\normalize\\dataset-normalize'
output_dir = 'C:\\Bangkit\\ML\\code\\preprocessing\\normalize\\new-dataset-normalize'
```

```python
process_directory(input_dir, output_dir)

jpeg_count = 0
total_count = 0
```

```python
for root, dirs, files in os.walk(output_dir):
    for file in files:
        total_count += 1
        if file.lower().endswith(".jpg"):
            jpeg_count += 1

print(f"Total images: {total_count}")
print(f"JPEG images: {jpeg_count}")
```
